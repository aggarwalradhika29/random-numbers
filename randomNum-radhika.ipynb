# Generating Random Numbers from Scratch:

```python
import time

def random_numbers(min,max, list):
    temp = str(time.perf_counter()) # assigns temp a string generated by the float value that is to be returned by perf_counter() & for every call, this values changes
    random = float(temp[::-1][:3:])/1000 # random variable is given a value from this particular index of the returned string in temp
    list.append((min+random*(max-min))) # a single random number is generated between min & max

min= int(input("Enter lower bound for range: "))
max= int(input("Enter upper bound for range: "))
n= int(input("How many numbers do you want to generate: "))
list1= []
for i in range (0,n): # for multiple calls
    random_numbers(min,max, list1);
print("Random numbers:\n",list1)
```

# Value of pi:

```python
list2= []
for i in range (0,n): # for multiple calls
    random_numbers(min,max, list2);
print(list2)
```

```python
from math import sqrt
# n is the total number of points
i= 0 #number of grid squares or points in the circle for random numbers
for i in range(n):
    x= list1[i]
    y= list2[i]
    r= sqrt(x**2 + y**2) #radius of the circle
    if r<1:
        i+=1
print("Value of pi: ",4*i/n)
```
